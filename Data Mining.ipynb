```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to extract product details from a category page
def extract_product_details(url, category):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
        "Accept-Language": "en-US, en;q=0.5"
    }
    
    response = requests.get(url, headers=headers)
    
    # Check if the request was successful
    if response.status_code != 200:
        print(f"Failed to retrieve data from {url}. Status code: {response.status_code}")
        return []
    
    soup = BeautifulSoup(response.content, 'html.parser')
    
    products = []
    for product in soup.find_all('div', {'data-component-type': 's-search-result'}):
        try:
            title = product.h2.get_text(strip=True) if product.h2 else 'N/A'
            
            # Extracting the price
            price_whole = product.find('span', 'a-price-whole').get_text(strip=True) if product.find('span', 'a-price-whole') else 'N/A'
            price_fraction = product.find('span', 'a-price-fraction').get_text(strip=True) if product.find('span', 'a-price-fraction') else '00'
            price=(price_whole)+(price_fraction)
            
            # Parsing price to float for comparison
            try:
                price_float = float(price.replace(',', '')) if price != 'N/A' else 'N/A'
            except ValueError:
                price_float = 'N/A'
            
            rating_str = product.find('span', {'class': 'a-icon-alt'}).get_text(strip=True) if product.find('span', {'class': 'a-icon-alt'}) else 'N/A'
            rating = float(rating_str.split()[0]) if rating_str != 'N/A' else 'N/A'
            
            review_count_str = product.find('span', {'class': 'a-size-base'}).get_text(strip=True) if product.find('span', {'class': 'a-size-base'}) else 'N/A'
            try:
                review_count = int(review_count_str.replace(',', ''))
            except ValueError:
                review_count = 'N/A'
            
            products.append({
                'Category': category,
                'Title': title,
                'Price': price_float,
                'Rating': rating,
                'Review Count': review_count,
            })
        except AttributeError as e:
            print(f"Attribute error while processing product: {e}")
        except Exception as e:
            print(f"Error occurred while extracting product details: {e}")
    
    return products

# URLs of category pages (replace with actual URLs for each category)
category_urls1 = {
    'Electronics': 'https://www.amazon.com/s?k=electronics',
    'Appliances': 'https://www.amazon.com/s?k=appliances',
    'Fashion': 'https://www.amazon.com/s?k=fashion',
    'Home': 'https://www.amazon.com/s?k=home',
    'Kitchen': 'https://www.amazon.com/s?k=kitchen',
}
category_urls2 = {
    'Health': 'https://www.amazon.com/s?k=health',
    'Sports': 'https://www.amazon.com/s?k=sports',
    'Fitness': 'https://www.amazon.com/s?k=fitness',
    'Baby Products': 'https://www.amazon.com/s?k=baby+products',
    'Books': 'https://www.amazon.com/s?k=books'
}

# Dictionary to store products by category
all_products = []

# Loop through each category and extract product details
for category, url in category_urls1.items():
    print(f"Extracting products for category: {category}")
    products = extract_product_details(url, category)
    all_products.extend(products)
```

```python
for category, url in category_urls2.items():
    print(f"Extracting products for category: {category}")
    products = extract_product_details(url, category)
    all_products.extend(products)
```

```python
# Convert the list of products to a pandas DataFrame and save to a CSV file
df = pd.DataFrame(all_products)
df.to_csv('amazon_products_prices.csv', index=False)

# Print the extracted product details
print(df)
```
